### convert the date columntype to date time type

```python
# convert the date column to date time 
f_export.Date=pd.to_datetime(f_export.Date)
f_gasoline_import.Date=pd.to_datetime(f_gasoline_import.Date)
f_gasoline_stocks.Date=pd.to_datetime(f_gasoline_stocks.Date)
f_production.Date=pd.to_datetime(f_production.Date)
f_supply.Date=pd.to_datetime(f_supply.Date)
gasoline_d_p.Date=pd.to_datetime(gasoline_d_p.Date)
gasoline_price.Date=pd.to_datetime(gasoline_price.Date)
gasoline_supply.Date=pd.to_datetime(gasoline_supply.Date)
reg=pd.to_datetime(reg.Date)
```

###  merge all the data into one data frame

```python
# merge the data into one data frame 
df=pd.merge(f_gasoline_stocks,gasoline_d_p,how='inner', on='Date')
df1=pd.merge(df,f_gasoline_import,how='inner', on='Date')
df2=pd.merge(df1,gasoline_supply,how='inner', on='Date')
df3=pd.merge(df2,gasoline_price,how='inner', on='Date')
```

```python
df4=pd.merge(df3,f_supply,how='inner', on='Date')
df5=pd.merge(df4,f_production,how='inner', on='Date')
df6=pd.merge(df5,f_export,how='inner', on='Date')
```

### Replace the nan values with 0

```python
# replace the nan values with 0 
df6['Weekly U.S. Exports of Finished Motor Gasoline  (Thousand Barrels per Day)']=df6['Weekly U.S. Exports of Finished Motor Gasoline  (Thousand Barrels per Day)'].replace(np.nan,0)
df6['Weekly U.S. Imports of Finished Motor Gasoline  (Thousand Barrels per Day)']=df6['Weekly U.S. Imports of Finished Motor Gasoline  (Thousand Barrels per Day)'].replace(np.nan,0)
```
